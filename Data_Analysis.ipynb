## Task 2:
## Cuisine Combination
### Identify the most common combinations of cuisines in the dataset

```python
# Convert non-string values in 'Cuisines' column to strings
data['Cuisines'] = data['Cuisines'].astype(str)

# Split the cuisines string into a list of cuisines
data['Cuisines'] = data['Cuisines'].str.split(', ')

# Create a list of all unique cuisines
unique_cuisines = data['Cuisines'].explode().unique()

# Create a new DataFrame with binary columns for each cuisine
cuisine_data = pd.DataFrame(0, index=data.index, columns=unique_cuisines)

# Assign 1 to the binary columns where the respective cuisine is present
for cuisine in unique_cuisines:
    cuisine_data[cuisine] = data['Cuisines'].apply(lambda x: 1 if cuisine in x else 0)

# Concatenate the original DataFrame with the new cuisine DataFrame
data_with_cuisines = pd.concat([data, cuisine_data], axis=1)

# Convert lists in 'Cuisines' column to tuples
data_with_cuisines['Cuisines'] = data_with_cuisines['Cuisines'].apply(tuple)

# Identify the most common combinations of cuisines
common_cuisine_combinations = data_with_cuisines.groupby('Cuisines').size().sort_values(ascending=False).head(10)

print("Top 10 most common combinations of cuisines:")
print(common_cuisine_combinations)
```

## LEVEL 3 : Task: Restaurant Reviews
## TASK 1 : - Analyze the text reviews to identify the most common positive and negative keywords.

```python
import pandas as pd
from nltk.tokenize import word_tokenize
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from collections import Counter

# Tokenize the text reviews and perform sentiment analysis
sid =SentimentIntensityAnalyzer()

positive_keywords = []
negative_keywords = []

for review in data['Rating text']:
    tokens = word_tokenize(review.lower())  # Convert to lowercase for consistency
    sentiment_score = sid.polarity_scores(review)['compound']  # Get sentiment score
    
    # Determine if the review is positive or negative based on sentiment score
    if sentiment_score >= 0.05:
        positive_keywords.extend(tokens)
    elif sentiment_score <= -0.05:
        negative_keywords.extend(tokens)

# Count the frequency of positive and negative keywords
positive_keyword_counts = Counter(positive_keywords)
negative_keyword_counts = Counter(negative_keywords)

# Get the most common positive and negative keywords
most_common_positive_keywords = positive_keyword_counts.most_common(5)
most_common_negative_keywords = negative_keyword_counts.most_common(5)

# Print the results
print("Most common positive keywords:")
for keyword, count in most_common_positive_keywords:
    print(keyword, count)

print("\nMost common negative keywords:")
for keyword, count in most_common_negative_keywords:
    print(keyword, count)
```

## Calculate the average length of reviews and explore if there is a relationship between review length and rating.


```python
# Calculate the length of each review and add it as a new column
data['Review Length'] = data['Rating text'].apply(lambda x: len(x))

# Calculate the average length of reviews
average_review_length = data['Review Length'].mean()

# Explore the relationship between review length and rating (correlation)
correlation = data['Review Length'].corr(data['Aggregate rating'])

# Print the results
print("Average length of reviews:", average_review_length)
print("Correlation between review length and rating:", correlation)
```

## TASK 3: Price Range vs. Online Delivery and Table Booking
## Analyze if there is a relationship between the price range and the availability of online delivery and table booking

```python
# Calculate the percentage of restaurants within each price range that offer online delivery and table booking
price_range_stats = data.groupby('Price range').agg({
    'Has Online delivery': lambda x: (x == 'Yes').mean() * 100,
    'Has Table booking': lambda x: (x == 'Yes').mean() * 100
}).reset_index()

# Rename columns for clarity
price_range_stats.rename(columns={'Has Online delivery': 'Online Delivery (%)', 'Has Table booking': 'Table Booking (%)'}, inplace=True)

# Print the results
print("Percentage of restaurants offering online delivery and table booking by price range:")
print(price_range_stats)
```

Consider the above result:- 
                         The relationship between the price range and online delivery is negatively correlated.
                         The relationship between the price range and Table Booking is positively correlated

##  Determine if higher-priced restaurants are more likely to offer these services.

```python
# Convert 'Has Online delivery' and 'Has Table booking' columns to binary values (0 for No, 1 for Yes)
data['Has Online delivery'] = (data['Has Online delivery'] == 'Yes').astype(int)
data['Has Table booking'] = (data['Has Table booking'] == 'Yes').astype(int)

# Calculate the average availability of online delivery and table booking for each price range
price_range_availability = data.groupby('Price range').agg({
    'Has Online delivery': 'mean',
    'Has Table booking': 'mean'
}).reset_index()

# Print the results
print("Average availability of online delivery and table booking by price range:")
print(price_range_availability)
```
